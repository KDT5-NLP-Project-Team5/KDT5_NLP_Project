```python
# 학습 데이터 처리
from soynlp import DoublespaceLineCorpus    # 한 개로 통합된 문서 데이터 분리
from soynlp.word import WordExtractor       # 단어 추출
```

```python
# 훈련 데이터 문서 분리

file_name = '../DATA/king_3.csv'

corpus = DoublespaceLineCorpus(file_name)
print(f'훈련 데이터 문서 : {len(corpus)}개')
```

```python
# 학습 진행
word_extractor = WordExtractor()

# 학습 진행하며 단어별 점수
word_extractor.train(corpus)

# 단어별 점수 확인
word_score_table = word_extractor.extract()
```

```python
# 단어별 점수표 확인
for idx, key in enumerate(word_score_table.keys()):
    print(f'[{idx}] - {key}')
    if idx == 10: break
```

### L Tokenizer 

```python
from soynlp.tokenizer import LTokenizer

# 토큰으로 쪼개기 위한 L토큰 기준값
scores = {word:score.cohesion_forward for word, score in word_score_table.items()} # cohesion_forward : 단어가 이어져 나올 확률

l_tokenizer = LTokenizer(scores=scores)
```

```python

# 파일 하나를 읽어와서 토큰화
with open('../DATA/king/세종.txt', 'r', encoding='utf-8') as f:
    text = f.read(5000)

result = l_tokenizer.tokenize(text)
for idx, token in enumerate(result):
    # 한 번에 10개씩 출력
    if idx % 20 == 0:
        print()
    print(token, end=' ')
```

```python
# 불용어 처리
from soynlp.normalizer import *
from soynlp.tokenizer import RegexTokenizer # RegexTokenizer : 정규표현식을 이용한 토큰화

tokenizer = RegexTokenizer()
token = tokenizer.tokenize(text)

# 명사 추출
from soynlp.noun import LRNounExtractor_v2

noun_extractor = LRNounExtractor_v2(verbose=False)
nouns = noun_extractor.train_extract(corpus)

print(f'명사 추출 결과 : {len(nouns)}개')
# print(nouns)

print(f'토큰화 결과 : {len(token)}개')
print(token)
```

### 데이터프레임으로 다루기

```python
import pandas as pd

df = pd.read_csv('../DATA/태조.txt', sep='\t', encoding='utf-8', header=None)
df
```

```python
# 데이터프레임 생성
df = pd.DataFrame(columns=['date', 'text', 'label'])
print(df)

file_name = '../DATA/태조.txt'

file = open(file_name, 'r', encoding='utf-8')
text = file.read()
file.close()
```

```python

# text에서 0 인덱스 출력

label = 0

for i, t in enumerate(text.split('\n')):
    index = i // 3
    if i % 3 == 1:
        # print(t)
        if index >= len(df):
            df.loc[index] = [None, None, label]
        df.iloc[index, 1] = t
    
    elif i % 3 == 0:
        # print('날짜 :', t)
        # dataframe에 추가
        if index >= len(df):
            df.loc[index] = [None, None, label]
        df.iloc[index, 0] = t
    
df['label'] = df['label'].astype(int)
df[:-1]
```

```python
kaa = df[:-1]

# 데이터프레임 정렬 date
kaa = kaa.sort_values(by='date')

# date를 splt()하여 년, 월, 일로 나누어 저장, 태조는 제거
kaa['date'][0].split(' ')[1:]

# 열을 추가하여 저장
kaa['year'] = kaa['date'].apply(lambda x: x.split(' ')[1])
kaa['month'] = kaa['date'].apply(lambda x: x.split(' ')[2])
kaa['day'] = kaa['date'].apply(lambda x: x.split(' ')[3])

# date 제거하고 label, year, month, day, text 순으로 저장
kaa = kaa[['label', 'year', 'month', 'day', 'text']]
kaa
```

```python
# csv로 저장
file = '../DATA/king_0.csv'
kaa.to_csv(file, index=False, encoding='utf-8', sep=';')
```
