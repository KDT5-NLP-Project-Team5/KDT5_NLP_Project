#### 먼저 도착한 파일들을 전처리하는 함수 만들기

구성
1. 파일 읽기, (.) 기준으로 문장 나누기
2. 전처리 : 정규 표현식으로 한글만 남기기
3. 인코딩, 임베딩 : 정수화, 패딩, 임베딩

```python
# 1. 파일 읽기 (폴더별 파일 읽기)
from os import listdir
from os.path import isfile, join
import pandas as pd

# 폴더 내 파일 목록 가져오기
path = '../DATA/txtt/'
files = [f for f in listdir(path) if isfile(join(path, f))]
# print(files)

# 파일 읽기
data = []
for file in files:
    with open(path + file, 'r', encoding='utf-8') as f:
        data.append(f.read())
print(data)
```

```python
# 토크나이저 Okt
from konlpy.tag import Okt
okt = Okt()

# 토큰화
token = []
for doc in data:
    token.append(okt.morphs(doc))
print(token)
```

```python
# 불용어 사전 불러오기
import re

with open('../DATA/stopwords.txt', 'r', encoding='utf-8') as f:
    stopwords = f.read().split()
print(stopwords)
```

```python
# 전처리 : def perprocess(data) 가져오기

# 기본적인 단어만 불용어 사전에 포함
stopwords = ['을', '를', '이', '가']

def preprocess(data, stopwords):
    # Okt
    okt = Okt()
    # 토큰화
    tokens = []
    for doc in data:
        # print(doc)
        tokens.append(okt.morphs(doc))
    
    # 【 문자 전까지만 추출
    
    
    # 정규표현식으로 한글과 . , " ' 만 추출
    tokens = [[token for token in doc if re.match(r'^[가-힣.,\'\"]+$', token)] for doc in tokens]
    
    # 불용어 제거
    tokens = [[token for token in doc if token not in stopwords] for doc in tokens]
    

    
    return tokens

# 15글자씩 끊어서 출력
for doc in preprocess(data, stopwords):
    for i in range(0, len(doc), 15):
        print(doc[i:i+15])
```
