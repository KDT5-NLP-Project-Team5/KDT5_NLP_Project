```python
import pandas as pd
```

```python
# 태조 csv
king0 = pd.read_csv('josuncsv/king_0.csv', sep=';')
king0
```

```python
# 따옴표가 있는 행만 출력
filtered_df = king0[king0['text'].str.contains('"')]
filtered_df
```

```python
import re

# 따옴표 정규화
def extract_dialogues(text):
    dialogue_pattern = r'".*?"'
    dialogues = re.findall(dialogue_pattern, text)
    return dialogues

filtered_df['dialogues'] = filtered_df['text'].apply(extract_dialogues)
filtered_df.dialogues
```

```python
# 각 행에 대화문 2개 이상인 경우 확인
filtered_df.dialogues.iloc[3]
```

```python
def extract_last_phrase(text):
    sentences = re.split(r'[.!?]', text)
    last_phrases = [sentence.split()[-1] for sentence in sentences if sentence.strip()]
    return last_phrases

# 마지막 어절 추출 및 새 컬럼 생성
filtered_df['last_phrases'] = filtered_df['text'].apply(extract_last_phrase)

filtered_df.last_phrases.iloc[3]
```
